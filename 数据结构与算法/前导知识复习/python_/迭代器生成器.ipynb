```python
list1: List[int] = [1, 2, 3, 4]
# 违法 TypeError: 'list' object is not an iterator
# print(next(l))

iter_l = iter(list1)
print(next(iter_l))  # 惰性, 需要下一个值时才返回
print("-" * 23)
while 1:
    # Return the next item from the iterator. If default is given and the iterator is exhausted, it is returned instead of raising StopIteration.
    # 从迭代器返回下一项。如果给定了default并且迭代器已耗尽，则返回它而不是引发StopIteration。
    try:
        print(next(iter_l))
    except StopIteration:
        print("没有下一个元素")
        break
```

```python
def fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b
for i in enumerate(fibonacci(),start=1):
    if i[0] == 1:
        print(i)
        break
```
